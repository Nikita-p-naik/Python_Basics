**1. What is the difference between a function and a method in Python?**

**Ans-**

**Function**
*	A function is independent and not tied to any object.
*	It is defined using the def keyword.
*	Can be used on its own.

Example-

def greet(name):

    return name
print(greet("Nikita"))

**Method**
*	A method is a function that belongs to an object (especially a class).
*	Called using the (.) notation.
*	Defined inside a class and always takes self as the first parameter (for instance methods).

Example:
class Person:

    def __init__(self, name):
        self.name = name

    def greet(self):  
        return f"Hello, {self.name}!"


p = Person("Nikita")

print(p.greet())

**3. What are the different ways to define and call a function in Python?**

```python
# 1. Regular Function (with def)
def greet(name):
    return name
print(greet("Nikita"))

#2. Function with Default Argument
def greet(name="Guest"):
    return f"Hello, {name}!"
print(greet())           # Output: Hello, Guest!
print(greet("Nikita"))   # Output: Hello, Nikita!

# 3. Function with Keyword Arguments
def profile(name, age):
    return f"{name} is {age} years old"
print(profile(age=25, name="Nikita"))


#4. Function with Variable-Length Arguments
#*args (non-keyword arguments):
def total(*numbers):
    return sum(numbers)
print(total(1, 2, 3))

#**kwargs (keyword arguments):

def show_details(**info):
    return info
print(show_details(name="Nikita", age=25))

#5. Lambda (Anonymous) Function
add = lambda a, b: a + b
print(add(2, 3))
```

**6. Explain the concept of generators in Python and how they are defined?**

**Ans-**
A generator is a special type of iterator that yields values one at a time using the yield keyword instead of return. it is memory efficient and automatically creates an iterator.

def count_up_to(n):

    i = 1
    while i <= n:
        yield i   
        i += 1

gen = count_up_to(3)

print(next(gen))

**9. Explain the purpose and usage of the `map()` function in Python.**

**Ans-**The map() function applies a function to each item in an iterable (like a list) and returns a new iterator with the results.
* Syntax:
map(function, iterable)
* Example:
nums = [1, 2, 3, 4]

print(squared)

**How it works:**
* Takes each element from the iterable (nums)

* Applies the function (lambda x: x**2)

* Returns a new iterable (needs to be converted to list() or tuple() to view)

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

for i in range(1, 201):
    if prime(i):
        print(i)
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fib:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            fib_num = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return fib_num
        else:
            raise StopIteration

fib = Fib(10)
for num in fib:
    print(num)

```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def pow_two(n):
    for i in range(n + 1):
        yield 2 ** i

for power in pow_two(4):
    print(power)
```

```python
# Step 1: Create and write to the file
with open("example.txt", "w") as f:
    f.write("Hello\n")
    f.write("This is a test file.\n")
    f.write("Line 3\n")

# Step 2: Define the generator function
def read_file_lines(filepath):
    with open(filepath, 'r') as file:
        for line in file:
            yield line.strip()

# Step 3: Use the generator to read the file line by line
for line in read_file_lines("example.txt"):
    print(line)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def rem_vow(txt):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, txt))

input_str = "I am A succesful Data Scientist"
result = rem_vow(input_str)
print(result)
```

```python
#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- â‚¬ if the value of the order is smaller than 100,00 â‚¬.
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "EinfÃ¼hrung in Python3, Bernd Klein", 3, 24.99]
]

def compute_order_totals(order_list):
    result = []
    for order in order_list:
        order_number = order[0]
        total = order[2] * order[3]  # quantity Ã— price
        if total < 100:
            total += 10  # add â‚¬10 if total < â‚¬100
        result.append((order_number, round(total, 2)))
    return result

final_totals = compute_order_totals(orders)
print(final_totals)
```
